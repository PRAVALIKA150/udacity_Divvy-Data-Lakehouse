```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Divvy Lakehouse Bronze").getOrCreate()
bronze_path = "hive_metastore.default"
```

```python
trips_df = spark.table("hive_metastore.default.trips_1_csv")
payments_df = spark.table("hive_metastore.default.payments_csv")
riders_df = spark.table("hive_metastore.default.riders_1_csv")
stations_df = spark.table("hive_metastore.default.stations_1_csv")
```

```python
display(trips_df)
```

```python
trips_df.write.format("delta").mode("overwrite").saveAsTable("trips")
payments_df.write.format("delta").mode("overwrite").saveAsTable("payments")
riders_df.write.format("delta").mode("overwrite").saveAsTable("riders")
stations_df.write.format("delta").mode("overwrite").saveAsTable("stations")
```

```python
spark.sql("CREATE DATABASE IF NOT EXISTS bronze")
spark.sql("CREATE TABLE IF NOT EXISTS bronze.trips USING DELTA LOCATION 'hive_metastore.default.trips'")
spark.sql("CREATE TABLE IF NOT EXISTS bronze.payments USING DELTA LOCATION 'hive_metastore.default.payments'")
spark.sql("CREATE TABLE IF NOT EXISTS bronze.riders USING DELTA LOCATION 'hive_metastore.default.riders'")
spark.sql("CREATE TABLE IF NOT EXISTS bronze.stations USING DELTA LOCATION 'hive_metastore.default.stations'")
```

```python
spark.sql("DROP TABLE IF EXISTS bronze.trips")
spark.sql("DROP TABLE IF EXISTS bronze.payments")
spark.sql("DROP TABLE IF EXISTS bronze.riders")
spark.sql("DROP TABLE IF EXISTS bronze.stations")
```
